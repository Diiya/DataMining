```python
#Creating data frames and reading each csv data file into variables.

df1 = pd.read_csv("data/test_data.csv")
df2 = pd.read_csv("data/train_data.csv")

print(df1.shape)
print(df2.shape)

print(df1)
print(df2)

```

```python
concat=df1+df2
print(concat)
print(concat.shape)
```

```python
type(concat)
```

```python
#Ref : https://cs230.stanford.edu/blog/split/
#Ref 2 for shuffle alone : https://www.geeksforgeeks.org/pandas-how-to-shuffle-a-dataframe-rows/

concat.sort_values(by=["Sentence_id"])  # make sure that the filenames have a fixed order before shuffling
random.seed(230)
data_1=concat.sample(frac = 1)
#random.shuffle(concat) # shuffles the ordering of filenames (deterministic given the chosen seed)

split1 = int(0.8 * len(data_1))
split2 = int(0.9 * len(data_1))
train_split = data_1[:split1]
dev_split= data_1[split1:split2]
test_split = data_1[split2:]
```

```python
print(train_split.shape)
print(test_split.shape)
print(dev_split.shape)
```
